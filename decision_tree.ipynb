```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Iris.csv')
```

```python
data
```

```python
data['species']=data['Species']
```

```python
data
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
data['species']=data['species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
data
```

```python
x=data.loc[:,('SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm')]
```

```python
y=data['Species']
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x,y)
```

```python
y
```

```python
plt.figure(figsize=(12, 8))
plot_tree(dt, filled=True, feature_names=x.columns, class_names=iris.target_names)
plt.show()
```
