```python
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data=pd.read_csv('Iris.csv')
```

```python
data
```

```python
new_data=data.drop('Species',axis='columns')
```

```python
new_data
```

```python
new_data.drop('Id',axis='columns',inplace=True)
```

```python
new_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_data=scaler.fit_transform(new_data)
```

```python
scaled_data
```

```python
data1=pd.DataFrame(scaled_data,columns=new_data.columns)
```

```python
data1
```

```python
list=[]
for i in range(1,10):
    model=KMeans(n_clusters=i,random_state=42).fit(data1)
    list.append(model.inertia_)
    
list
```

```python
series=pd.Series(list,index=range(1,10))
```

```python
import matplotlib.pyplot as plt
plt.plot(series)
plt.show
```

Let us take cluster number as 3

```python
kmean=KMeans(n_clusters=3)
```

```python
kmean.fit_transform(data1)
```

```python
cluster=kmean.predict(data1)
```

```python
from sklearn.decomposition import  PCA
pca=PCA(n_components=2)
```

```python
pca_d=pca.fit_transform(scaled_data)
```

```python
pcapca_data=pd.DataFrame(pca_d,columns=['c1','c2'])
```

```python
pcapca_data
```

```python
pcapca_data['cluster']=cluster
```

```python
pcapca_data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(pcapca_data['c1'],pcapca_data['c2'],c=pcapca_data['cluster'])
```
